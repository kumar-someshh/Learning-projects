# Preprocessing


1. Data Cleaning: Cleaning the data by removing duplicate, null values, handling missing vlaues and removing outliers.

2. Data Transformatiom: involves feature encoding, feature scaling.

3. Data reduction: selecting relevant features.

````python
# encoding "Fuel_Type" Column
data.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

# encoding "Seller_Type" Column
data.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

# encoding "Transmission" Column
data.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
````

````python
data.rename(columns = {'Selling_Price':'Selling_Price(lacs)','Present_Price':'Present_Price(lacs)','Owner':'Past_Owners'},inplace = True)
````

````python
data['Age'] = 2024 - data['Year']
data.drop('Year',axis=1,inplace = True)
data.drop('Car_Name', axis = 1, inplace =True)
````

**When to Use Each Technique:**

**Outlier Removal:** Consider removing outliers if:

You have strong evidence they are errors or irrelevant.
They are very rare occurrences and distort the data distribution.
Your dataset is large enough to handle the loss of data points.

**Feature Scaling:** Use feature scaling for most machine learning algorithms, especially those sensitive to feature scales, to:
Ensure equal contribution from all features.
Improve model convergence and potentially performance.

````python
y = data['Selling_Price(lacs)']
x = data.drop('Selling_Price(lacs)',axis=1)
````

````python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("x train: ",x_train.shape)
print("x test: ",x_test.shape)
print("y train: ",y_train.shape)
print("y test: ",y_test.shape)
````

Applying regression models

* Linear Regression
* Ridge Regression
* Lasso Regression

# **Linear Regression**

````python
from sklearn.linear_model import LinearRegression
model_one = LinearRegression()

# Training model
model_one.fit(x_train,y_train)


# Model evaluation using R^2 score
from sklearn.metrics import r2_score

# training set
y_pred_train = model_one.predict(x_train)
R2_train_model = r2_score(y_train,y_pred_train)
print("R^2 score for training set: ", R2_train_model)

# test set
y_pred_test = model_one.predict(x_test)
R2_test_model = r2_score(y_test,y_pred_test)
print("R^2 score for test set: ", R2_test_model)
````

````python



# Create the figure and subplots for a 1 row, 2 columns layout
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot training data on the first subplot (ax1)
ax1.scatter(y_train, y_pred_train, label='Training Data', c='blue')
ax1.set_xlabel("Actual Price")
ax1.set_ylabel("Predicted Price")
ax1.set_title("Training Data: Actual vs. Predicted Prices")
ax1.legend()

# Plot testing data on the second subplot (ax2)
ax2.scatter(y_test, y_pred_test, label='Testing Data', c='red')
ax2.set_xlabel("Actual Price")
ax2.set_ylabel("Predicted Price")
ax2.set_title("Testing Data: Actual vs. Predicted Prices")
ax2.legend()

# Adjust spacing between subplots for better readability (optional)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)  # Adjust values as needed

plt.show()
````
