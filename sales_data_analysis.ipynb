#Q.1
What was the best month for slaes? How much was earned that month?

```python
# Are there any missing values in the dataset?
# yes
data.isna().sum()
data.nunique()
```

```python
data.dropna(inplace=True)
```

```python
# lets augment another column for Months
# data[["Months", "Date"]] = data["Order Date"].str.split( "/" , n=1, expand = True)

# more easy way
data["Months"] =  data["Order Date"].str[0:2]
data
```

```python
# data["Months"].astype("int16")
# we are having trouble converting the dtype of "Months" column, lets find out the issue
data["Months"].unique()

# the problem is there are some ilogical values each column has, we found out that some string values are causing this problem.
```

```python
print(data["Product"].unique())
print(data["Quantity Ordered"].unique())
print(data["Price Each"].unique())


# yes there are some illogical vaues present in the dataset, lets find those rows
```

```python
clean_data = data.drop_duplicates( keep =False)
```

```python
clean_data
```

```python
clean_data["Months"].value_counts()
clean_data["Months"].astype("int")
# congrats you did it now its a good time to ponder about how would you deal with such outlier values if object is string object and uniques values are in thousands or more.
# well, each column or feature will have some rules of what type of value it can take, for example our months feature can take values from 1 to 12 in the same way we can
# define a structure, thus a way to found outliers.
```

```python
clean_data["Quantity Ordered"]=  pd.to_numeric(clean_data["Quantity Ordered"], downcast= "integer")
clean_data["Price Each"] = pd.to_numeric( clean_data["Price Each"], downcast = "float")
clean_data["Sales"]=clean_data["Quantity Ordered"]* clean_data["Price Each"]
```

```python
clean_data["Sales"]
```

```python
clean_data.info()
```

**let's answer our question**

``What was the best month for slaes? How much was earned that month?``


```python
# Assuming "Sales" is the column containing sales figures
sales_by_month = clean_data.groupby("Months")["Sales"].sum()
sales_by_month
# it's the month of descember with highest sale with a value of 4.6 million
```

```python
months = range(1,13)
plt.bar(months, sales_by_month)
plt.xlabel("Month number")
plt.ylabel("Sales in USD($)")
plt.xticks(months)
plt.show()
```

# Q.2
**which city had highest number of sales?**

```python
clean_data.head()
```

```python
# city = clean_data["Purchase Address"].str.split(",", expand = True)
# city = city.iloc[: , 1]
# clean_data["City"]= city
```

```python
def get_city(x):
  return x.split(",")[1]

def get_state(x):
  return x.split(",")[2].split(" ")[1]

clean_data["City"] = clean_data['Purchase Address'].apply(lambda x: get_city(x) + " (" + get_state(x) + ") ")
```

```python
clean_data.head()
clean_data["City"].unique()
```

```python
sales_per_city = clean_data.groupby("City")["Sales"].sum()
#  San Francisco    8247283.000
print(sales_per_city.index)
print(sales_per_city.values)
```

```python


# Create the bar chart
plt.bar(sales_per_city.index, sales_per_city.values)  # Use index for city names, values for sales

# Add labels and title
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.title('Sales per City in clean_data')

# Rotate x-axis labels if many cities (optional)
plt.xticks(rotation=45, ha='right')  # Rotate for readability with many cities

# Show the chart
plt.show()
```

# Q.4
**What products are most often sold together?**

```python
clean_data.head()
```

```python
clean_data
```

```python
# look at the id number 176560 at row index 3 and 4, they have the same order id means thy were ordered together.
# So, there are duplicate values of order id, lets group the Product names assoiciated with one single order id by adding another column named "Bought_together"

df = clean_data[clean_data['Order ID'].duplicated(keep=False)]
df['Bought_together'] = df.groupby('Order ID')["Product"].transform(lambda x: ','.join(x))

```

```python
df
```

```python
# lets drop the duplicated values
df = df[['Order ID', 'Bought_together']].drop_duplicates()
df.head()
```

```python
# @title Most Common Product Pairs (Pairs Only)

df[df['Bought_together'].str.count(',') == 1]['Bought_together'].value_counts().head(10).plot(kind='barh')
```
